```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

band_gaps = np.array(tcc["Perovskite_band_gap"])
pces = np.array(tcc["JV_default_PCE"])
print("已加载 tcc 数据。")
# 筛选 PCE <= 30 的数据
valid_pce_mask = pces <= 30
band_gaps = band_gaps[valid_pce_mask]
pces = pces[valid_pce_mask]
print(f"已移除 {np.sum(~valid_pce_mask)} 个 PCE > 30 的数据点。")
# 创建图表
plt.figure(figsize=(10, 6))

# 绘制所有数据点
sns.scatterplot(x=band_gaps, y=pces, label='All data points', alpha=0.6, s=50)

# 筛选 PCE > 17 的数据
high_pce_mask = pces > 17
high_pce_band_gaps = band_gaps[high_pce_mask]
high_pce_values = pces[high_pce_mask]

# 绘制 PCE > 17 的数据点，用不同颜色和标记突出显示
if len(high_pce_band_gaps) > 0:
    sns.scatterplot(x=high_pce_band_gaps, y=high_pce_values, color='red', s=100, label='PCE > 17%', marker='o', edgecolor='black')

    # 计算 PCE > 17 的数据对应的带隙的 10% 和 90% 分位数
    band_gap_10th_percentile = np.percentile(high_pce_band_gaps, 10)
    band_gap_90th_percentile = np.percentile(high_pce_band_gaps, 90)

    # 使用垂直阴影区域标记带隙的 10%-90% 分位数范围
    plt.axvspan(band_gap_10th_percentile, band_gap_90th_percentile, color='yellow', alpha=0.3,
                label=f'High-efficiency range({band_gap_10th_percentile:.2f}-{band_gap_90th_percentile:.2f} eV)')

    # 可选：添加垂直线标示分位数点
    plt.axvline(band_gap_10th_percentile, color='orange', linestyle='--', linewidth=1, label=f'10th percentile({band_gap_10th_percentile:.2f} eV)')
    plt.axvline(band_gap_90th_percentile, color='orange', linestyle='--', linewidth=1, label=f'90th percentile({band_gap_90th_percentile:.2f} eV)')


    print(f"PCE> 17% 的带隙 10% 分位数: {band_gap_10th_percentile:.2f} eV")
    print(f"PCE > 17% 的带隙 90% 分位数: {band_gap_90th_percentile:.2f} eV")
    print(f"带隙的 10%-90% 分位数范围大致在 {band_gap_10th_percentile:.2f} eV 到 {band_gap_90th_percentile:.2f} eV 之间。")


else:
    print("数据中没有 PCE > 17% 的点。")


# 添加标题和轴标签
# 添加标题和轴标签
plt.xlabel('Bandgap(eV)')
plt.ylabel('Power Conversion Efficiency (%)')
plt.title('Bandgap vs.Power Conversion Efficiency(PCE > 17%)')

# 设定 Y 轴范围
plt.ylim(0, 30)
# 调整图例位置防止重叠，并显示网格线
plt.legend(loc='best') # 尝试自动选择最佳位置
plt.grid(True, linestyle='--', alpha=0.6)

# 保存图表到当前路径
plt.savefig('bandgap_vs_pce.png', dpi=300, bbox_inches='tight')
print("图表已保存为 'bandgap_vs_pce.png'")

# 显示图表
plt.show()
```

```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('lastceshi.csv')

# 保存第一列的原始名称
first_column_name = df.columns[0]

# 创建一个映射字典（从当前列名到新列名）
current_to_new_names = {
    r'$CA_m$': 'Cell_area_measured',
    r'$CA_a$': 'Cell_architecture',
    r'$PBG$': 'Perovskite_band_gap',
    r'$PD_n$': 'Perovskite_deposition_number_of_deposition_steps',
    r'$PD_q$': 'Perovskite_deposition_quenching_induced_crystallisation',
    r'$C_{nocp}$': 'Cell_number_of_cells_per_substrate',
    r'$C_{flb}$': 'Cell_flexible',
    r'$C_{smtp}$': 'Cell_semitransparent',
    r'$PC_{ABC3}$': 'Perovskite_composition_perovskite_ABC3_structure',
    r'$PC_{lf}$': 'Perovskite_composition_leadfree',
    r'$PD_{sa}$': 'Perovskite_deposition_solvent_annealing',
    r'$AL_{front}$': 'Add_lay_front',
    r'$JV_{light}$': 'JV_light_intensity',
    r'$JV_{masked}$': 'JV_light_masked_cell',
    r'$ETL_t$': 'ETL_thickness_zong',
    r'$SP_m$': 'Stabilised_performance_measured',
    r'$A_{MN}$': 'a_ionsMendeleevNumber',
    r'$A_{AW}$': 'a_ionsAtomicWeight',
    r'$a_{MT}$': 'a_ionsMeltingT',
    r'$C_{norm}$': '0-norm',
    r'$LF_{MN}$': 'long_formMendeleevNumber',
    r'$LF_{AW}$': 'long_formAtomicWeight',
    r'$LF_{MT}$': 'long_formMeltingT',
    r'$LF_{E}$': 'long_formElectronegativity',
    r'$LF_{V}$': 'long_formNfValence',
    r'$CF_b$': 'Cell_flexible',
    r'$CS_p$': 'Cell_semitransparent'
}

# 如果第一列的名称在映射字典中，将其移除
if first_column_name in current_to_new_names:
    del current_to_new_names[first_column_name]

# 替换列名（除了第一列）
df = df.rename(columns=current_to_new_names)

# 保存更新后的数据到CSV文件
df.to_csv('lastceshi_renamed.csv', index=False)

print("列名已成功替换，第一列保持不变，新文件已保存为'lastceshi_renamed.csv'")
```
